```python
import pandas as pd 
df =pd.read_csv('C://Data_science/Pandas/cross_tab.csv')
df
```

```python
pd.crosstab(df.Nationality,df.Hand)
```

```python
pd.crosstab(df.Sex,df.Hand)
```

```python
pd.crosstab(df.Hand,df.Nationality)
```

```python
pd.crosstab(df.Hand,df.Nationality,margins=True) #Gives sum of Rows and Columns 
```

```python
pd.crosstab(df.Sex,[df.Hand,df.Nationality],margins=True)
```

```python
pd.crosstab([df.Nationality,df.Sex],[df.Hand],margins=True)
```

```python
pd.crosstab([df.Sex],[df.Hand],margins=True)
```

```python
pd.crosstab([df.Sex],[df.Hand],normalize='index')
```

```python
import numpy as np 
pd.crosstab([df.Sex],[df.Hand],values=df.Age,aggfunc=np.average)
```
